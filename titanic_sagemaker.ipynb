```python
import pandas as pd
import numpy as np

df=pd.read_csv('titanic_training.csv',header=None)
df
```

```python
import sagemaker
from sagemaker import get_execution_role

role=get_execution_role()
```

```python
binary_estimator=sagemaker.LinearLearner(role=role,
                                        train_instance_count=1,
                                        train_instance_type='ml.m4.xlarge',
                                        predictor_type='binary_classifier',
                                        num_classes=2)
```

```python
mini_batch_size=100
records=df.values.astype('float32')
labels=df[0].astype('float32')
train_records=binary_estimator.record_set(records,labels,channel='train')
binary_estimator.fit([train_records])
```

```python
binary_predictor=binary_estimator.deploy(initial_instance_count=1,
                                        instance_type='ml.t2.medium')
```
